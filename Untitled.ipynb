```python
hvi = pd.read_csv("hvi-nta-2020.csv")
nta_url = "https://data.cityofnewyork.us/resource/9nt8-h7nd.geojson?$limit=5000"
nta = geo.read_file(nta_url)

print(nta.columns)
```

```python
# Rename NTA boundary code to match HVI file
nta = nta.rename(columns={"nta2020": "NTACode"})

# Merge HVI data with NTA boundaries
nta_hvi = nta.merge(
    hvi,
    on="NTACode",
    how="left"
)

nta_hvi[["NTACode", "HVI_RANK"]].head()
```

```python
# Load fountain data
fountains = pd.read_csv(
    "Cool_It!_NYC_2020_-_Drinking_Fountains_20260716.csv"
)

# Remove commas and convert coordinates to numbers
fountains["x"] = fountains["x"].str.replace(",", "").astype(float)
fountains["y"] = fountains["y"].str.replace(",", "").astype(float)

# Convert fountains into spatial points
fountains_geo = geo.GeoDataFrame(
    fountains,
    geometry=geo.points_from_xy(
        fountains["x"],
        fountains["y"]
    ),
    crs="EPSG:2263"
)

# Match NTA map projection
fountains_geo = fountains_geo.to_crs(nta_hvi.crs)

# Check
fountains_geo.head()
```

```python
from matplotlib.lines import Line2D

fig, ax = plt.subplots(figsize=(11, 11))

# Plot areas without HVI data
nta_hvi[nta_hvi["HVI_RANK"].isna()].plot(
    ax=ax,
    color="#D9D9D9",
    edgecolor="#AAAAAA",
    linewidth=0.15
)

# Plot HVI areas
hvi_plot = nta_hvi[nta_hvi["HVI_RANK"].notna()].plot(
    column="HVI_RANK",
    cmap="YlOrRd",
    linewidth=0.15,
    edgecolor="#AAAAAA",
    legend=True,
    legend_kwds={
        "label": "Heat Vulnerability Rank",
        "orientation": "horizontal",
        "shrink": 0.7,
        "pad": 0.02
    },
    ax=ax
)

# Move the colorbar to the bottom
colorbar = ax.get_figure().axes[-1]
colorbar.set_position([
    0.25,   # left
    0.05,   # bottom
    0.5,    # width
    0.025   # height
])

# Plot fountains
fountains_geo.plot(
    ax=ax,
    color="#1f77b4",
    markersize=4,
    alpha=0.75
)

# Create fountain legend
fountain_marker = Line2D(
    [0],
    [0],
    marker='o',
    color='w',
    label='Drinking Fountain',
    markerfacecolor='#1f77b4',
    markersize=4
)

ax.legend(
    handles=[fountain_marker],
    loc="lower right",
    frameon=True
)

# Formatting
ax.set_title(
    "NYC Heat Vulnerability Index and Drinking Fountains",
    fontsize=16
)

ax.axis("off")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(11,11))

# Areas without greenspace data
nta_hvi[nta_hvi["GREENSPACE"].isna()].plot(
    ax=ax,
    color="#D9D9D9",
    edgecolor="#AAAAAA",
    linewidth=0.15
)

# Greenspace map
nta_hvi[nta_hvi["GREENSPACE"].notna()].plot(
    column="GREENSPACE",
    cmap="Greens",
    linewidth=0.15,
    edgecolor="#AAAAAA",
    legend=True,
    legend_kwds={
        "label": "Greenspace Score",
        "orientation": "horizontal",
        "shrink": 0.7,
        "pad": 0.02
    },
    ax=ax
)

ax.set_title(
    "NYC Greenspace Availability by Neighborhood",
    fontsize=16
)

ax.axis("off")

plt.tight_layout()
plt.show()
```
